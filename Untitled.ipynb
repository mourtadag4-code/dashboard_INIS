```python
import pandas as pd
import os
import shutil
from datetime import datetime

# ============================================================
# CONFIGURATION
# ============================================================
BASE_FILE = "donnees_avicole_bd.xlsx"
SOURCE_FILE = "Controle Ferme (1).xlsx"
BACKUP_FILE = f"donnees_avicole_bd_backup_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx"

# ============================================================
# 1. SAUVEGARDER LA BASE ACTUELLE
# ============================================================
if os.path.exists(BASE_FILE):
    shutil.copy2(BASE_FILE, BACKUP_FILE)
    print(f"✅ Base sauvegardée : {BACKUP_FILE}")
else:
    print(f"⚠️ Fichier {BASE_FILE} non trouvé")

# ============================================================
# 2. LIRE LA BASE ACTUELLE
# ============================================================
print("\n📂 Lecture de la base actuelle...")
df_ventes = pd.read_excel(BASE_FILE, sheet_name="ventes")
df_journalier = pd.read_excel(BASE_FILE, sheet_name="journalier")
df_recap = pd.read_excel(BASE_FILE, sheet_name="cycles_recap")

if "acheteur" not in df_ventes.columns:
    df_ventes["acheteur"] = ""
    print("   → Colonne 'acheteur' ajoutée")

print(f"   → {len(df_ventes)} lignes de ventes existantes")

# ============================================================
# 3. EXTRAIRE LES ACHETEURS DEPUIS L'ONGLET Rhytme de vente
# ============================================================
def extraire_acheteurs_rhytme_vente(fichier_source):
    """Extrait les acheteurs depuis l'onglet Rhytme de vente"""
    acheteurs_dict = {}
    
    # Lire l'onglet Rhytme de vente sans en-tête
    df_rhytme = pd.read_excel(fichier_source, sheet_name="Rhytme de vente", header=None)
    
    # Blocs de ventes par cycle (lignes de début)
    blocs = [
        {"cycle": "Cycle1", "start": 3, "end": 16},   # lignes 4 à 16 (index 3 à 15)
        {"cycle": "Cycle2", "start": 25, "end": 31},   # lignes 26 à 31 (index 25 à 30)
        {"cycle": "Cycle3", "start": 48, "end": 70}    # lignes 49 à 70 (index 48 à 69)
    ]
    
    for bloc in blocs:
        cycle = bloc["cycle"]
        print(f"\n   📖 Extraction pour {cycle} (lignes {bloc['start']+1} à {bloc['end']+1})")
        
        for idx in range(bloc["start"], bloc["end"] + 1):
            if idx >= len(df_rhytme):
                break
                
            row = df_rhytme.iloc[idx]
            
            # Structure des colonnes (0-indexé)
            # B = index 1 (Jour), C = index 2 (Date), D = index 3 (Quantité), G = index 6 (Contact)
            jour = row.iloc[1] if pd.notna(row.iloc[1]) else None
            quantite = row.iloc[3] if pd.notna(row.iloc[3]) else None
            acheteur = row.iloc[6] if pd.notna(row.iloc[6]) else None
            
            # Vérifier que c'est une ligne de vente valide
            if jour and quantite and quantite > 0 and acheteur and str(acheteur) != "nan":
                acheteur_str = str(acheteur).strip()
                print(f"      → J{jour} : {acheteur_str} ({quantite} sujets)")
                acheteurs_dict[(cycle, jour)] = acheteur_str
    
    return acheteurs_dict

# ============================================================
# 4. EXTRAIRE LES ACHETEURS
# ============================================================
print("\n📂 Extraction des acheteurs depuis Rhytme de vente...")
acheteurs_dict = extraire_acheteurs_rhytme_vente(SOURCE_FILE)

print(f"\n   → {len(acheteurs_dict)} associations (cycle, jour) → acheteur trouvées")

# ============================================================
# 5. METTRE À JOUR LA BASE AVEC LES ACHETEURS
# ============================================================
print("\n🔄 Mise à jour de la colonne 'acheteur'...")

compteur = 0
for idx, row in df_ventes.iterrows():
    key = (row["cycle_id"], row["jour"])
    if key in acheteurs_dict:
        ancien = df_ventes.loc[idx, "acheteur"]
        nouveau = acheteurs_dict[key]
        if ancien != nouveau:
            df_ventes.loc[idx, "acheteur"] = nouveau
            compteur += 1
            print(f"   → {key[0]} - J{key[1]} : '{ancien}' → '{nouveau}'")

print(f"\n   → {compteur} lignes mises à jour")

# ============================================================
# 6. SAUVEGARDER LA BASE MISE À JOUR
# ============================================================
print(f"\n💾 Sauvegarde de la base mise à jour...")
with pd.ExcelWriter(BASE_FILE, engine='openpyxl') as writer:
    df_journalier.to_excel(writer, sheet_name="journalier", index=False)
    df_recap.to_excel(writer, sheet_name="cycles_recap", index=False)
    df_ventes.to_excel(writer, sheet_name="ventes", index=False)

print(f"\n✅ Pipeline terminé ! Base mise à jour : {BASE_FILE}")

# ============================================================
# 7. RÉSUMÉ
# ============================================================
print("\n📊 RÉSUMÉ DES ACHETEURS PAR CYCLE")
for cycle in ["Cycle1", "Cycle2", "Cycle3"]:
    df_cycle = df_ventes[df_ventes["cycle_id"] == cycle]
    if not df_cycle.empty and "acheteur" in df_cycle.columns:
        top = df_cycle.groupby("acheteur")["quantite"].sum().sort_values(ascending=False).head(3)
        print(f"\n{cycle}:")
        for acheteur, qte in top.items():
            if acheteur and acheteur != "":
                print(f"   - {acheteur}: {qte:,.0f} sujets")
```
